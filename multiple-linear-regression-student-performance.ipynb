```python
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

# Load Dataset

```python
data=pd.read_csv('/kaggle/input/student-performance-multiple-linear-regression/Student_Performance.csv')
data.head()
```

```python
data.info()
```

# Encode Categoral Data

```python
le=LabelEncoder()
data['Extracurricular Activities']=le.fit_transform(data['Extracurricular Activities'])
data
```

```python
x=data.drop(columns=['Performance Index'])
x
```

```python
y=data['Performance Index']
y
```

# Train Dataset

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,train_size=0.8,random_state=101)
```

# Apply Linear Regresstion Model

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
y_predict
```

# Accuracy 

```python
r2Score=r2_score(y_test,y_predict)
print("Accuracy : ",r2Score)
```
